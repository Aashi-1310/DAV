```python
import pandas as pd
import numpy as np
```

```python
#QUES3: Create a dataframe having at least 3 columns and 50 rows to store numeric data generated using a random function. Replace 10% of the values by null values whose index positions are generated using random function.
df=pd.DataFrame(np.random.randint(0,100,size=(50,3)))
n=15
while(n):
    y=np.random.randint(1,50)
    x=np.random.randint(0,3)
    df.loc[y,x]=np.nan
    n=n-1
print(df)

```

```python
#a) Identify and count missing values in a dataframe.
x=df.isnull()
count=0
for i in range(50):
    for j in range(3):
        if (x.loc[i,j]):
            count+=1
            print("(", i, ",", j,")")
print("Total number of missing values are:", count)
```

```python
#f) Find the correlation between first and second column and covariance between second and third column.
print("The correlation between first and second column is: ")
print(df[0].corr(df[1]))
print("The covariance between second and third column is : ")
print(df[1].cov(df[2]))
```

```python
#h) Discretize second column and create 5 bins
#bins=[20,40,60,80,100]
cats=pd.cut(df[1],bins=5)
print(cats)
```

```python
#d) Sort the dataframe on the basis of the first column.
df.sort_values(by=[1])
```

```python
#b) Drop the column having more than 5 null values.
li=[]
x=df.isnull()
for i in range(0,3):
    count=0
    for j in range(50):
        if (x.loc[j,i]):
                count+=1
    li.append(count)
for m,k in enumerate(li):
    if(k>5):
        df=df.drop(df.columns[m],axis=1)
print(df)
```

```python
s=df.sum(axis=1)
print("The Sum is: ")
print(s)
print("the Maximum sum is: ", s.max())
row= df.sum(axis=1).idxmax()
print("Row having maximum sum is: ", row)
df=df.drop(row, axis=0)
print("New dataframe: ")
print(df)
```

```python
#e) Remove all duplicates from the first column.
df.drop_duplicates()
```

```python
import matplotlib.pyplot as plt
data_a= np.array([-900,1,2,3,4, 600, 5,6,900])
q1,q2,q3= np.percentile(sorted(data_a),[25,50,75])
print(sorted(data_a))
print("total =", data_a.size)
print("q1 =", q1)
print("q2 =", q2)
print("q3 =", q3)
iqr=q3-q1
print("iqr =", iqr)

lb= q1-(1.5* iqr)
ub= q3 + (1.5*iqr)

print("Lower bound", lb)
print("Upper bound", ub)
outliers= [i for i in data_a if i<= lb or i>=up]
print(outliers)
plt.boxplot(data_a)
plt.show()
    
```
